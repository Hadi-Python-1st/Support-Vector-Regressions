## **What IS Support Vector Regression**

Support Vector Regression (SVR) is a machine learning
technique used for regression tasks. It is a variant of
Support Vector Machines (SVM) and is designed to
predict continuous numeric values, making it suitable
for tasks like time series forecasting, stock price
prediction, and more.

## **Implementing SVR on Height vs weight dataset**

1. **Importing libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

2. **Load The Dataset**

```python
file_path = "HeightVsWeight.csv"

data = pd.read_csv(file_path)

features = ["Age"]
label = ["Height"]

x = data.loc[:, features].values
y = data.loc[:, label].values
```

3. **Training Testing and splitting**

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2,random_state=1)
```

4. **Feature Scalling and Normalizing the Data**

```python
from sklearn.preprocessing import StandardScaler
sc_x = StandardScaler()
sc_y = StandardScaler()
x_train = sc_x.fit_transform(x_train)
x_test = sc_x.transform(x_test)
y_train = sc_y.fit_transform(y_train)
# y_test = sc_y.transform(y_test)
```

5. **Fitting the Model**

```python
from sklearn.svm import SVR
reg = SVR(kernel = 'rbf')
reg.fit(x_train, y_train)
```

6. **Plotting the Graph**

```python
plt.scatter(x_train, y_train, color = 'red')
plt.plot(x_train, reg.predict(x_train), color = "blue")
plt.title('Height vs Weight (Training set)')
plt.xlabel('Age')
plt.ylabel('Height')
plt.show()
```

7. **Comparing Results**

```python
np.set_printoptions(precision=2) # For Printing only upto 2 decimal places
y_pred = sc_y.inverse_transform(reg.predict(x_test).reshape(-1, 1)) # Reshape the predictions to a 2D array
print(np.concatenate((y_test.reshape(len(y_test) , 1) , y_pred.reshape(len(y_pred) , 1)) , 1))
```
